```python
# データ処理
import numpy as np
import pandas as pd 

# 学習モデル
from xgboost import XGBRFClassifier

# ハイパーパラメータチューニング
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from bayes_opt import bayesian_optimization
import optuna
```

# サンプルデータの読み込み

```python
from sklearn.datasets import load_breast_cancer

# アヤメデータのインスタンスを生成
cancer = load_breast_cancer() 

feature = pd.DataFrame(cancer.data, columns=cancer.feature_names) # 特徴量
target = pd.DataFrame(cancer.target, columns=["target"]) #　ターゲット

df = pd.concat([feature, target], axis=1) # 特徴量とターゲットの結合

display(df)
```

# train validation test split

```python
from sklearn.model_selection import train_test_split

# train/ validation split
X_train, X_valid, y_train, y_valid = train_test_split(feature,
                                                      target,
                                                      test_size=0.3,
                                                      shuffle=True,
                                                      stratify=True,
                                                      random_state = 1234)
```
