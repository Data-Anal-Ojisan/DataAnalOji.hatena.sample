```python
# データ処理
import numpy as np
import pandas as pd 

# 学習モデル
from xgboost import XGBClassifier

# 分類問題評価指標
from sklearn.metrics import classification_report

# ハイパーパラメータチューニング
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from bayes_opt import bayesian_optimization
import optuna

import warnings
warnings.filterwarnings('ignore')
```

# サンプルデータの読み込み

```python
from sklearn.datasets import load_breast_cancer

# 癌データのインスタンスを生成
cancer = load_breast_cancer() 

feature = pd.DataFrame(cancer.data, columns=cancer.feature_names) # 特徴量
target = pd.DataFrame(cancer.target, columns=["target"]) #　ターゲット

df = pd.concat([feature, target], axis=1) # 特徴量とターゲットの結合

display(df)
```

# train validation test split

```python
from sklearn.model_selection import train_test_split

# train split
X_train, X_tmp, y_train, y_tmp = train_test_split(feature,
                                                  target,
                                                  test_size=0.3,
                                                  shuffle=True,
                                                  stratify=target,
                                                  random_state=1234)

# validation test split
X_valid, X_test, y_valid, y_test = train_test_split(X_tmp,
                                                    y_tmp,
                                                    test_size=0.5,
                                                    shuffle=True,
                                                    stratify=y_tmp,
                                                    random_state=1234)

del X_tmp, y_tmp
```

# デフォルトパラメータ

```python
# インスタンスの生成
xgb_without_tuning = XGBClassifier(random_state=1234)

# 学習
xgb_without_tuning.fit(X_train, y_train)

# 学習後の分類精度
display(pd.DataFrame(classification_report(y_true=y_test,
                                           y_pred=xgb_without_tuning.predict(X_test),
                                           target_names=cancer.target_names,
                                           output_dict=True)))
```
